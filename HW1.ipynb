```python
print("---------------1--------------")
import numpy as np
a = np.array([[1, 6], 
              [2, 8],
              [3, 11],
              [3, 10],
              [1, 7]])
mean_a = a.mean(axis=0)
print("Средние значения столбцов: {}".format(mean_a))

print("---------------2--------------")
a_centered = a - mean_a
print(a_centered)
```

```python
print("---------------3--------------")
a_centered_sp = a_centered[:,0] @ a_centered[:,1]
print(a_centered[0:])
print(f"a_centered_sp = {a_centered_sp}")
print(f"a_centered_sp / N-1 = {a_centered_sp / (a.shape[0] - 1)}")
```

```python
print("---------------4--------------")
#print(a)
print(f"Значение ковариации = {np.cov(a.T)[0,1]}")
```

```python
### Тема “Работа с данными в Pandas”
import pandas as pd
print("---------------1--------------")

authors = pd.DataFrame({
    'author_id': [1, 2, 3],
    'author_name': ['Тургенев', 'Чехов', 'Островский'],
})

authors

```

```python

books = pd.DataFrame({
    'author_id': [1, 1, 1, 2, 2, 3, 3], 
    'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
    'price': [450, 300, 350, 500, 450, 370, 290]
})
books
```

```python
authors_price = pd.merge(authors, books, on='author_id', how='inner')
authors_price
```

```python
authors_price.sort_values(by="price", inplace=True, ascending = False)
top5 = authors_price[0:5]
top5
```

```python
authors_stat = authors_price.loc[:, ['author_name', 'price']]
authors_stat["min_price"] = authors_price["price"]
authors_stat["max_price"] = authors_price["price"]
authors_stat["mean_price"] = authors_price["price"]

authors_stat

```

```python
groupby = authors_stat.groupby("author_name")
groupby.agg({"min_price": "min", "max_price": "max", "mean_price": "mean"})
```

```python
authors_price['cover']=['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
authors_price
```

```python
 book_info = pd.pivot_table(authors_price, values='price', index=['author_name'], columns=['cover'], aggfunc=np.sum)
```
